```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# delete unnamed column there is no use of this
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize = (5,5))
ax= sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

```python
#from the above chart the numnber of females is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("relationship between parents education and students score")
plt.show()
```

```python
#from the above chart  we have concluded that the education of the parents have a good impact in their scores

```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("relationship between parents maritial status and students score")
plt.show()
```

```python
#from the above chart we have analyze that its neglisible impact in the numbers due to the maritial status
```

```python
#DISTRIBUTION OF ETHNIC GROUPS
```

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist =[groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
